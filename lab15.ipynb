```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
#1-1
df = iris_x

df1 = df.dropna()
display(df1.head(3))
```

```python
# NaN이 2개 이상인 행 삭제

#1-2
df_thresh = df.dropna(thresh = 2)
display(df_thresh.head(10))
```

```python
# 1열, 2열에 NaN 있는 경우 삭제

#1-3
df_na = df.dropna(subset=['sepal_length', 'sepal_width'], how='any')
display(df_na.head(10))
```

```python
# NaN을 0으로 치환

#1-4
df_zero = df.fillna(0)
display(df_zero.head(10))
```

```python
# 누락 데이터를 평균값으로 치환

#1-5
df_mean = df.fillna(round(df.mean(),1))
display(df_mean.head(10))                    
```

```python
# NaN값을 10으로 치환

#1-6
df_ten = df.fillna(10)
display(df_ten.head(10))
```

```python
# NaN 값을 윗 행 값으로 대체

#1-7
df_up = df.fillna(method='ffill')
display(df_up.head(10))
```

```python
#3-1
df = pd.read_csv('./data/product_click_new.log', header = None, sep = ' ')
print(df.head(5))
print(df.info())
```

```python
#3-2
df.columns = ['oldclicktime', 'clickpid']
print(df.head(5))
```

```python
#3-3
df['oldclicktime'] = df['oldclicktime'].astype('str')
df.info()
```

```python
#3-4
df['newclicktime'] = pd.to_datetime(df['oldclicktime'])
print(df.info())
print(df.head(5))
```

```python
#3-5
df['year'] = df['newclicktime'].dt.year
df['month'] = df['newclicktime'].dt.month
df['day'] = df['newclicktime'].dt.day
df['hour'] = df['newclicktime'].dt.hour
df['minute'] = df['newclicktime'].dt.minute

print(df.info())
display(df.head(5))
```
