```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
print(font_name)
rc('font', family=font_name)
```

```python
#1
df = pd.DataFrame({'1분기':[500, 690, 1100, 1500, 1990, 1020], 
                   '2분기':[450, 700, 1030, 1650, 2020, 1600],
                   '3분기':[520, 820, 1200, 1700, 2300, 2200], 
                   '4분기':[610, 900, 1380, 1850, 2420, 2550]},
                  index=['2015', '2016', '2017', '2018', '2019', '2020'])
display(df)
```

```python
#2
df.index = df.index.map(lambda x : str(x) + '년')
plt.figure(figsize=(10,6))
plt.title('시각화 과제1', size=20, c='purple')
plt.plot(df)
```

```python
#3
df1 = df.T
plt.figure(figsize=(10,6))
plt.title('시각화 과제2', c='red', size=20)
plt.plot(df1)
```

```python
#4
colors = ['red', 'green', 'brown', 'purple']
explode = [0.05, 0.05, 0.05, 0.05]

plt.figure(figsize=(10,6))
plt.pie(df.sum(), labels=df.columns, explode=explode, colors=colors, autopct='%.1f%%', shadow=True)
plt.show()
```

```python
#5
plt.figure(figsize=(10,6))
plt.pie(df1.sum(), labels=df1.columns, autopct='%.1f%%')
plt.show()
```
